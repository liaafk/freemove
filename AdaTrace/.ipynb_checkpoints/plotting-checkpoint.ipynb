```python
import matplotlib.pyplot as plt
```

```python
def plot_adatrace(filename, title):
    """
    plots trajectories from AdaTrace-formatted file
    """
    with open(filename) as fi:
        traces = []
        for ln in fi:
            if ln.startswith(">"):
                traces.append(ln[3:])
    plt.title(title)
    for trace in traces:
        coord = trace.split(';')[:-1]
        x = [float(i.partition(',')[0]) for i in coord]
        y = [float(i.partition(',')[-1]) for i in coord]
        plt.plot(x, y)
    plt.show()
```

```python
plot_adatrace('vru_adatrace.dat', 'Real dataset')
plot_adatrace('vru_adatrace.dat-eps1.0-iteration2.dat', "Synthetic dataset with total epsilon=1 no attacks")
plot_adatrace('vru_adatrace.dat-eps2.0-iteration2.dat', "Synthetic dataset with total epsilon=2 no attacks")
plot_adatrace('vru_adatrace.dat-eps4.0-iteration2.dat', "Synthetic dataset with total epsilon=4 no attacks")
```
