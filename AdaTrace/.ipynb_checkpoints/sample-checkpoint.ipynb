```python
import random
random.seed(42)
```

```python
with open("brinkhoff.dat") as myfile:
    lines = myfile.readlines()
whole = [lines[i]+lines[i+1] for i in range(0, len(lines), 2)]
```

```python
whole[0]
```

```python
random.seed(42)
for i in range(10):
    sample = random.sample(whole, 2000)
    with open("brinkhoff_2000_"+str(i)+".dat", "w") as text_file:
        text_file.write(''.join(e for e in sample))
    with open("brinkhoff_1000_"+str(i)+".dat", "w") as text_file:
        text_file.write(''.join(e for e in sample[:1000]))
```
