```python
import numpy as np
import os
```

```python
data_location = 'Geolife Trajectories 1.3/Data'
```

```python
users = [str(item).zfill(3) for item in np.arange(182)]
```

```python
users
```

```python
for user in users:
    if not os.path.exists('transformed_geolife/'+user):
        os.makedirs('transformed_geolife/'+user)
```

```python
for user in users:
    traj_dir = data_location+'/'+user+'/Trajectory'
    traj_files = [f for f in os.listdir(traj_dir) if os.path.isfile(os.path.join(traj_dir, f))]
    user_traces = []
    print("reading from files")
    for traj_file in traj_files:
        file = open(traj_dir+'/'+traj_file)
        coordinates = [line.split(',')[:2] for line in file.readlines()[6:]]
        user_traces.append(coordinates)
        file.close()
    file = open('transformed_geolife/'+user+'.dat',"a")
    i = 0
    print("writing to file")
    for trace in user_traces:
        to_write = "#"+str(i)+":\n>0:"
        for point in trace:
            to_write += point[0]+','+point[1]+';'
        to_write += '\n'
        file.write(to_write)
        i += 1
    file.close()
    print(user)
```
