## what do?
### 1. delete first line
### 2. change (measurement) ID to filename
### 3. delete 5th column
### 4. put em all together in a huge ass .csv file
### 5. normalize time (so that time >= 0)

```python
import fileinput
import os
import glob
import pandas as pd
```

```python
for filename in os.listdir(os.getcwd()):
    if filename.endswith("txt"):
        with open(filename, "r") as f:
            lines = f.readlines()
            new_content = ""
            for i in range(1, len(lines)):
                values = lines[i].split()
                #change id to filename
                values[0] = filename.replace(".txt", "").replace("_","")
                #delete 5th row
                new_content = new_content+",".join(values[0:4])+"\n"
            f.close()
        with open(filename, "w") as f:
            f.write(new_content)
            f.close()
```

```python
all_filenames = [i for i in glob.glob('*.{}'.format('txt'))]
for filename in all_filenames:
    with open(filename, 'r') as original: data = original.read()
    with open(filename, 'w') as modified: modified.write("id,timestamp,x,y\n" + data)
    read_file = pd.read_csv(filename)
    read_file.to_csv(filename.replace('txt', 'csv'), index=False, encoding='utf-8-sig')
all_filenames = [i for i in glob.glob('*.{}'.format('csv'))]
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ]).sort_values([])
print(combined_csv)
```

```python
combined_csv.to_csv( "trajectories.csv", index=False, encoding='utf-8-sig')
```
